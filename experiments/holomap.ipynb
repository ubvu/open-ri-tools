Reproducable example of HoloMap pane update delay

```python
import panel as pn
import hvplot.networkx as hvnx
import networkx as nx
from holoviews import HoloMap, DynamicMap
from holoviews.util import Dynamic
```

```python
# number of hvplots
n = 20
```

```python
# attributes for alpha
attributes = {str(i):i/n for i in range(n)}
```

```python
# create graph
G = nx.Graph()
G.add_edge('a', 'b', **attributes)
pos = nx.spring_layout(G)
```

HoloMap

```python
# create hvplots
hvplots = {}
for i in attributes:
    nodes = hvnx.draw_networkx_nodes(G, pos)
    edges = hvnx.draw_networkx_edges(G, pos, alpha=i)
    hvplots[int(i)] = nodes * edges
    
# create HoloMap
hm = HoloMap(hvplots)
```

```python
hm
```

```python
pn.panel(hm)
```

HoloMap -> DynamicMap

```python
dhm = Dynamic(hm)
```

```python
pn.panel(dhm)
```

Dynamic map

```python
def make_plot(alpha_attr):
    nodes = hvnx.draw_networkx_nodes(G, pos)
    edges = hvnx.draw_networkx_edges(G, pos, alpha=str(alpha_attr))
    return nodes * edges
    
# create HoloMap
dm = DynamicMap(make_plot, kdims='alpha').redim.values(alpha=[int(i) for i in attributes])
```

```python
dm
```

```python
pn.panel(dm)
```
